## **Pareto Chart for Hospital Operational Inefficiencies**

### Problem Statement

Hospitals face multiple operational challenges impacting efficiency and patient care, but not all issues contribute equally. Identifying the most significant causes is essential for prioritizing improvements.

### Why Pareto Chart?

Prioritization of Causes: Shows which few causes contribute most (the “vital few”) to hospital inefficiencies versus the “trivial many.”
- **Visual Impact:** Combines bar and cumulative line charts for clear understanding of impact distribution.
- **Data-Driven Decisions:** Enables hospital management to focus resources on solving key problems with highest impact.
- **Supports Continuous Monitoring:** Can be updated regularly to track progress and emerging issues.


```python
import matplotlib.pyplot as plt
import numpy as np

causes = [
    'Staff Shortages During Busiest Days',
    'Manual Billing Errors in IPD',
    'Lack of Demand Forecasting',
    'Limited Analytics-Driven Decisions',
    'Seasonal Patient Volume Spikes'
]
impact = [30, 25, 20, 15, 5]
cumulative = np.cumsum(impact)

# Plot
fig, ax1 = plt.subplots(figsize=(12, 10))  # Adjust size as needed
bars = ax1.bar(causes, impact, color='skyblue', label='Impact (%)')

# Bar labels
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height + 1, f'{height}%', ha='center', fontsize=9)

# Axis labels and ticks
ax1.set_xlabel('Causes of Hospital Inefficiencies', fontsize=12)
ax1.set_ylabel('Impact (%)', fontsize=12, color='blue')
ax1.tick_params(axis='y', labelcolor='blue')

# X-tick fix (rotate and wrap)
plt.xticks(rotation=45, ha='right', fontsize=9)

# Cumulative line
ax2 = ax1.twinx()
ax2.plot(causes, cumulative, color='red', marker='o', label='Cumulative (%)')
ax2.set_ylabel('Cumulative (%)', fontsize=12, color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title and legends
plt.title('Pareto Chart: Causes of Hospital Operational Inefficiencies', fontsize=14)
ax1.legend(loc='upper left')
ax2.legend(loc='upper right')

# Fix layout and spacing
plt.tight_layout()
plt.savefig('pareto_hospital_inefficiencies.png', dpi=150)
plt.show()
```


# Use Case

By analyzing causes such as staff shortages during peak days, billing errors, and lack of demand forecasting, the hospital uses a Pareto chart to reveal that 55% of inefficiencies come from just two causes: staff shortages and manual billing errors. 

This insight helps:

Allocate budget to improve staffing and training on busiest days
Automate billing to reduce errors

Implement forecasting tools for patient volume
This targeted approach significantly enhances operational efficiency and patient satisfaction.

## Fishbone Diagram for Hospital Operational Inefficiencies

Problem Statement

Hospitals struggle with multiple operational challenges such as understaffing, inventory shortages, inefficient processes, and fluctuating patient volumes. These issues cause delays, increased costs, reduced patient satisfaction, and staff burnout.

Why Fishbone Diagram?

Structured Root Cause Analysis: Fishbone diagrams visually organize potential causes into categories, making complex problems easier to analyze.

Comprehensive View: It captures multiple factors (people, process, equipment, environment) contributing to inefficiencies.

Facilitates Team Collaboration: Helps different hospital departments identify and agree on problem sources together.

Prioritizes Solutions: By clearly showing causes, it aids in targeting the most critical issues for intervention.


```python
import matplotlib.pyplot as plt

# Create figure and axis
fig, ax = plt.subplots(figsize=(30, 20))

# Draw the main spine (arrow)
ax.arrow(0.1, 0.5, 0.8, 0, head_width=0.05, head_length=0.05, fc='blue', ec='blue')
ax.text(0.9, 0.46, 'Hospital Operational Inefficiencies', fontsize=20, ha='center')

# Define categories and positions
categories = ['People', 'Process', 'Equipment', 'Environment']
positions = [0.2, 0.4, 0.6, 0.8]

# Draw bones and labels
for pos, category in zip(positions, categories):
    ax.plot([pos, pos - 0.05], [0.5, 0.65], 'k-')  # Top bone
    ax.plot([pos, pos - 0.05], [0.5, 0.35], 'k-')  # Bottom bone
    ax.text(pos - 0.07, 0.67, category, fontsize=20, ha='right')

# Causes per category
causes = {
    'People': ['Understaffing on Busiest Days', 'Overstaffing on Quiet Days', 'Lack of Staff Training'],
    'Process': ['High OPD Wait Times', 'IPD Billing Errors'],
    'Equipment': ['Inventory Stock-Outs', 'LAB Equipment Downtime', 'Insufficient Equipment for Peaks'],
    'Environment': ['Patient Volume Spikes', 'Fluctuating Demand Patterns'],
}

# Place causes on diagram
for pos, category in zip(positions, categories):
    for i, cause in enumerate(causes[category]):
        y_offset = 0.65 - (i * 0.05) if i % 2 == 0 else 0.35 + (i * 0.05)
        ax.text(pos - 0.1, y_offset, cause, fontsize=20, ha='right')

# Hide axis details
ax.set_xticks([])
ax.set_yticks([])
ax.axis('off')

plt.savefig('fishbone_hospital_inefficiencies.png', bbox_inches='tight')
plt.show()
```

This leads to smoother operations and better patient care.


## ICU Staff Scheduling Optimization
Problem Statement
Hospitals must ensure adequate ICU staff coverage every day. Scheduling should minimize total active staff to control costs, while maintaining at least a minimum number of staff on duty for patient safety and care quality.
## Solution Method:

Linear Integer Optimization using Binary Decision Variables (0 = not scheduled, 1 = scheduled). Solved using PuLP.


## What is Linear Programming (LP)?

Linear Programming is a method to find the best solution to a problem with limited resources. It helps decide how to do something optimally, like minimizing costs or maximizing efficiency, using simple math formulas.

## Why Linear Programming?

Optimal Resource Utilization: Linear programming finds the minimal number of staff needed to meet coverage requirements.

Clear Constraints Handling: Easily enforce minimum coverage and staff availability constraints (e.g., on leave).

Binary Decisions: Staff are either scheduled (1) or not (0), fitting naturally into binary variable models.

Cost Efficiency: Helps reduce overstaffing and associated costs without risking patient care.

also: 
- To **make the best decisions** based on data.  
- To **use resources efficiently** (like staff, money, or time).  
- To handle **problems with clear rules and limits**.  
- It is **easy to understand and solve** using computers.  
- Useful for many real-world tasks like scheduling and budgeting.


## How Does Linear Programming Work?

1. **Decide the variables:** What you want to find out (e.g., who to schedule).  
2. **Set the goal:** What you want to achieve (e.g., minimize staff numbers).  
3. **Add constraints:** Rules that must be followed (e.g., at least 2 staff needed).  
4. **Solve the problem:** Use software to get the best solution that fits the goal and rules.

**In short:** LP helps you make smart choices with limited resources by using simple math and computers.



```python
from pulp import LpProblem, LpMinimize, LpVariable, lpSum

staff_df = dataframes['staff scheuling']
icu_staff = staff_df[staff_df['Department'] == 'ICU']
icu_staff = icu_staff[icu_staff['Is On Leave?'] == 'No']

# Decision variables
staff_vars = {row['Name']: LpVariable(row['Name'], cat='Binary') for _, row in icu_staff.iterrows()}

# Problem: Minimize total staff while ensuring min. coverage
prob = LpProblem("ICU_Staff_Coverage", LpMinimize)
prob += lpSum(staff_vars.values())  # minimize total active staff
prob += lpSum(staff_vars.values()) >= 2  # at least 2 ICU staff needed

prob.solve()
icu_schedule = pd.DataFrame([
    {"Name": name, "Scheduled": int(var.value())} for name, var in staff_vars.items()
])
print(icu_schedule)
print(icu_schedule['Scheduled'].value_counts())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create DataFrame
df = pd.DataFrame(icu_schedule)

# Convert 1/0 to "Yes"/"No" for readability (optional, if needed in the data)
df['Scheduled_Label'] = df['Scheduled'].apply(lambda x: 'Yes' if x == 1 else 'No')

# Plot
plt.figure(figsize=(8, 4))
bars = plt.bar(df['Name'], df['Scheduled'])

# Annotate with "Yes"/"No" on top of bars
for bar, label in zip(bars, df['Scheduled_Label']):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() + 0.05,
             label, ha='center', va='bottom', fontsize=10)

# Title and labels
plt.title('Scheduling Status')
plt.ylabel('Scheduled')
plt.xticks(rotation=45)
plt.yticks([0, 1], ['No', 'Yes'])
plt.ylim(0, 1.3)

# Legend
# Use fake (invisible) bars to make a legend
green_bar = plt.bar([], [], color='blue', label='Scheduled: Yes')
red_bar = plt.bar([], [], color='red', label='Scheduled: No')
plt.legend()

# Show plot
plt.tight_layout()
plt.show()
```

## Inventory Reordering Optimization
Problem Statement:

How to minimize the risk of stockouts while reducing excess inventory costs?

Solution Method:

Simple prescriptive logic: Reorder quantity = max(0, threshold - current quantity). Multiply by unit price to get total reorder cost.

```python
inventory=pd.read_excel('Weekly_Hospital_Inventory_Updated.xlsx')
inventory['Reorder Threshold'] = 500
inventory['Reorder Qty'] = inventory['Reorder Threshold'] - inventory['Quantity']
inventory['Reorder Qty'] = inventory['Reorder Qty'].apply(lambda x: x if x > 0 else 0)
inventory['Reorder Cost'] = inventory['Reorder Qty'] * inventory['Unit Price (INR)']
inventory[['Item Name', 'Category', 'Reorder Qty', 'Reorder Cost']]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming your dataframe is named 'inventory'
# Sample columns: 'Item Name', 'Category', 'Reorder Qty', 'Reorder Cost'

# Group reorder quantities and costs by Category
category_summary = inventory.groupby('Category').agg({
    'Reorder Qty': 'sum',
    'Reorder Cost': 'sum'
}).reset_index()

# Plot 1: Total Reorder Quantity by Category
plt.figure(figsize=(10,5))
sns.barplot(data=category_summary, x='Category', y='Reorder Qty', palette='viridis')
plt.title('Total Reorder Quantity by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Plot 2: Total Reorder Cost by Category
plt.figure(figsize=(10,5))
sns.barplot(data=category_summary, x='Category', y='Reorder Cost', palette='magma')
plt.title('Total Reorder Cost by Category')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Optional: Top 10 Items by Reorder Cost
top_items = inventory.sort_values(by='Reorder Cost', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.barplot(data=top_items, x='Item Name', y='Reorder Cost', palette='coolwarm')
plt.title('Top 10 Items by Reorder Cost')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

Cost Control: Avoids overstocking and reduces unnecessary spending by reordering only what's needed.

Prevents Stockouts: Ensures essential items are always available, avoiding treatment delays.
